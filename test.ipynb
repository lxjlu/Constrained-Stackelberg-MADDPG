```python
reward_leader = reward[0]
mean_count = 100
length = len(reward_leader)-mean_count
episodes = np.arange(mean_count, length)
mean_reward = np.zeros(length-mean_count)
for i in range(length-mean_count):
    # rate[i] = np.mean(success[i:i+mean_count])
    mean_reward[i] = np.mean(reward_leader[i:i+mean_count])
plt.plot(mean_reward)
plt.xlabel("Episode")
plt.ylabel("Reward of leader agent")
```

```python
crash = np.load(file_path+'crash_record.npy')
mean_count = 100
length = len(crash)-mean_count
episodes = np.arange(mean_count, length)
crash_rate = np.zeros(length-mean_count)
for i in range(length-mean_count):
    # rate[i] = np.mean(success[i:i+mean_count])
    crash_rate[i] = np.sum(crash[i:i+mean_count])/mean_count
plt.ylim(0, 1)
plt.xlim(50, 700)
plt.xlabel('Episode',{'size':16})
plt.ylabel('Rate', {'size':16})
plt.tick_params(labelsize=14)
# plt.fill_between(episodes, 0, non_target_means, color='wheat')
plt.fill_between(episodes, 0, 1-crash_rate, color='lightgreen')
plt.fill_between(episodes, 1-crash_rate, 1, color='lightpink')

plt.legend(['safe', 'crash'], loc=4)
```

```python
leader_arrive = np.load(file_path+'leader_arrive_record.npy')
follower_arrive = np.load(file_path+'follower_arrive_record.npy')
leader_arrive = np.array(leader_arrive, dtype=int)
follower_arrive = np.array(follower_arrive, dtype=int)
mean_count = 50
length = len(leader_arrive)-mean_count
episodes = np.arange(mean_count, length)
leader_rate = np.zeros(length-mean_count)
follower_rate = np.zeros(length-mean_count)
fail_rate = np.zeros(length-mean_count)
for i in range(length-mean_count):
    # rate[i] = np.mean(success[i:i+mean_count])
    leader_rate[i] = np.sum(leader_arrive[i:i+mean_count])/mean_count
    follower_rate[i] = np.sum(follower_arrive[i:i+mean_count])/mean_count
    fail_rate[i] = 1-leader_rate[i]-follower_rate[i]
plt.ylim(0, 1)
plt.xlim(50, 350)
plt.xlabel('Episode',{'size':16})
plt.ylabel('Rate', {'size':16})
plt.tick_params(labelsize=14)
# plt.fill_between(episodes, 0, non_target_means, color='wheat')
plt.fill_between(episodes, 0, leader_rate, color='lightgreen')
plt.fill_between(episodes, leader_rate, follower_rate+leader_rate, color='yellow')
plt.fill_between(episodes, leader_rate+follower_rate, 1, color='lightpink')

plt.legend(['leader', 'follower','crash'], loc=4)
```
